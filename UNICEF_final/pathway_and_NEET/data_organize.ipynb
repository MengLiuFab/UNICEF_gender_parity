# load the percentage data


```python
data = pd.read_excel(datafile)
group_list = data['group'].unique().tolist()
print('all poluation group are :')
print('\n'.join(group_list))

category_list = data['small_categ'].unique().tolist()
print('\nall small categories are :')
print('\n'.join(category_list))
```

```python

group_level = data[data['group']==group_list[0]]
cat_level = group_level[group_level['small_categ']==category_list[0]]
DF_percentage = cat_level[['country','source','year','upsec_ageindatayear','source']]

# here we take the first group to initial the dataframe.
for group in group_list:
    for category in category_list:
        group_level = data[data['group']==group]
        cat_level = group_level[group_level['small_categ']==category]
        tab = f'{group}_{category}'
        DF_percentage[tab] = cat_level['percentage'].values

DF_percentage = DF_percentage.reset_index().drop('index',axis='columns')
# remove the regions such as world. 
DF_percentage = DF_percentage.drop(DF_percentage[pd.isna(DF_percentage['year'])].index)    
# split the age range into specific age     
DF_percentage['age_low'] = DF_percentage['upsec_ageindatayear'].str.split('-').str[0].astype('int')
DF_percentage['age_high'] = DF_percentage['upsec_ageindatayear'].str.split('-').str[1].astype('int')
DF_percentage['age_band'] = DF_percentage['age_high'] - DF_percentage['age_low'] + 1
for group in group_list:
    DF_percentage[f'{group}_Needs_in_primary'] = DF_percentage[f'{group}_Never entered primary'] + DF_percentage[f'{group}_Dropped out of primary'] + DF_percentage[f'{group}_Still attending primary'] 
    DF_percentage[f'{group}_Needs_in_LS'] = DF_percentage[f'{group}_Did not transition to lower secondary'] + DF_percentage[f'{group}_Dropped out of lower secondary'] + DF_percentage[f'{group}_Still attending lower secondary']
    DF_percentage[f'{group}_Needs_in_US'] = DF_percentage[f'{group}_Did not transition to upper secondary']
display(DF_percentage)
#DF_percentage = DF_percentage[['country','year','upsec_ageindatayear','Female_Needs_in_primary','Female_Needs_in_LS','Female_Needs_in_US','Male_Needs_in_primary','Male_Needs_in_LS','Male_Needs_in_US','age_low','age_high','age_band']]
DF_percentage


tmp = DF_percentage[DF_percentage['country']=='Liberia']

tmp[['Male_Did not transition to upper secondary','Female_Did not transition to upper secondary']]
```

# revise the NEET code, directly devide the neet data between the two gender 


```python
dfNEET = pd.read_excel('EIP_NEET_SEX_AGE_NB_A_EN.xlsx',skiprows = [0,1,2,3,4])
countries = dfNEET['Reference area'].unique().tolist()
genders = ['Female','Male']
data = dict()
data['country'] = countries
for gender in genders:
    percenlist_low = []
    percenlist_high = []
    yearlist = []
    for country in countries:
        tmp = dfNEET[dfNEET['Reference area']==country]
        tmp_gender = tmp[tmp.Sex==gender].iloc[-1,:]
        percenlist_low.append(tmp_gender['15-19'])
        percenlist_high.append(tmp_gender['20-24'])
        yearlist.append(tmp_gender['Time'])
    data[f'{gender}_NEET_15_19'] = percenlist_low
    data[f'{gender}_NEET_20_24'] = percenlist_high
    data['year'] = yearlist
df_percentage = pd.DataFrame(data)
df_percentage
```

```python
df_percentage['GPI_15_19'] = df_percentage['Female_NEET_15_19'] / df_percentage['Male_NEET_15_19']
df_percentage['GPI_20_24'] = df_percentage['Female_NEET_20_24'] / df_percentage['Male_NEET_20_24']
data_NEET_final = df_percentage[['country','GPI_15_19','GPI_20_24']]
```

```python
data_final = data_path_final
data_final = data_final.merge(data_NEET_final,how='right')
display(data_final)
display(data_path_final)

data_final.rename(
    {
        'Gender_parity_index_primary':'GPI_path_primary',
        'Gender_parity_index_LS':'GPI_path_LS',
        'Gender_parity_index_US':'GPI_path_US',
        'GPI_15_19':'GPI_NEET_15_19',
        'GPI_20_24':'GPI_NEET_20_24',        	   
    },
    axis=1,
    inplace=True
    
)
data_final.to_csv('ForAlexa.csv')

data_final


```

# cumulative method


```python
levels = ['Never entered primary','Did not transition to lower secondary','Did not transition to upper secondary']
genders = ['Female','Male']
level_item = levels = ['primary','LS','US']
for gender in genders:
    df[f'{gender}_primary_pop_cumulative'] = df[f'{gender}_Never entered primary_pop'] 
    df[f'{gender}_LS_pop_cumulative'] = df[f'{gender}_Did not transition to lower secondary_pop'] + df[f'{gender}_Never entered primary_pop']
    df[f'{gender}_US_pop_cumulative'] = df[f'{gender}_Did not transition to upper secondary_pop'] + df[f'{gender}_Did not transition to lower secondary_pop'] + df[f'{gender}_Never entered primary_pop']
for level in levels:
    df[f'GPI_{level}_cumulative'] = df[f'Female_{level}_pop_cumulative']/df[f'Male_{level}_pop_cumulative']
    
```

```python
df.replace([np.inf],np.nan,inplace=True)
df.replace([],np.nan,inplace=True)
```

```python
df = df.dropna(subset= ['GPI_primary_cumulative','GPI_LS_cumulative','GPI_US_cumulative'],how='all')
df
```
